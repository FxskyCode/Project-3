```python
def map_maker(place):
    coord = [place["latitude"],place["longitude"]]
    maps = Map(location = coord, zoom_start = 15, control_scale = True)
    myicon = Icon(color = "black", prefix = "fa", icon = "fa-solid fa-building")
    company = Marker(location = coord, tooltip = place["name"], icon = myicon)
    company.add_to(maps)
    myicon = Icon(color = "blue", prefix = "fa", icon = "fa-solid fa-plane")
    airport = Marker(location = [place["airport_lat"],place["airport_lon"]], tooltip = "airport", icon = myicon)
    airport.add_to(maps)
    myicon = Icon(color = "red", prefix = "fa", icon ="fa-graduation-cap")
    school = Marker(location = [place["school_lat"],place["school_lon"]], tooltip = "school", icon = myicon)
    school.add_to(maps)
    myicon = Icon(color = "purple", prefix = "fa", icon ="fa-moon-o")
    night= Marker(location = [place["n_club_lat"],place["n_club_lon"]], tooltip = "night_club", icon = myicon)
    night.add_to(maps)
    myicon = Icon(color = "orange", prefix = "fa", icon =" fa-futbol-o")
    stadium = Marker(location = [place["b_stadium_lat"],place["b_stadium_lon"]], tooltip = "night_club", icon = myicon)
    stadium.add_to(maps)
    myicon = Icon(color = "green", prefix = "fa", icon ="fa-coffee")
    bar = Marker(location = [place["starbucks_lat"],place["starbucks_lon"]], tooltip = "night_club", icon = myicon)
    bar.add_to(maps)
    
    return maps
```
